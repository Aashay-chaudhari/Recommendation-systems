#Create Recommendation Algorithm
#Content Based

```python
import numpy as np
import tensorflow as tf

# tf.reset_default_graph()
print(tf.__version__)
```

```python
users = user_ids
movies = movies
features = user_movie_columns

num_users = len(users)
num_movies = len(movies)
num_feats = len(features)
```

```python
user_movies = final_user_movie_matrix_np

movies_feats = final_movies_matrix_np
```

```python
users_movies = tf.constant(user_movies, dtype = tf.float32)
movies_feats = tf.constant(movies_feats, dtype = tf.float32)
```

```python
wgtd_feature_matrices=[tf.expand_dims(tf.transpose(users_movies)[:,i],axis=1)*movies_feats for i in range(num_users)]

users_movies_feats = tf.stack(wgtd_feature_matrices)
```

```python
users_movies_feats
```

```python
users_movies_feats_sums = tf.reduce_sum(users_movies_feats, axis = 1)
users_movies_feats_totals = tf.reduce_sum(users_movies_feats_sums, axis = 1)
```

```python
users_movies_feats
```

```python
users_movies_feats_sums
```

```python
users_movies_feats_totals
```

```python
user_feat = [users_movies_feats_sums[i, :]/ users_movies_feats_totals[i]for i in range(num_users)]
```

```python
user_feat
```

```python
users_rating = [tf.map_fn(lambda x: tf.tensordot(user_feat[i], x, axes = 1), movies_feats) for i in range(num_users)]

all_users_ratings = tf.stack(users_rating)
```

```python
all_users_ratings_new = tf.where(tf.equal(users_movies, tf.zeros_like(users_movies)),
                                 all_users_ratings,
                                 -np.inf*tf.ones_like(tf.cast(users_movies, tf.float32)))
```

```python
def find_user_top_movies(user_index, num_to_recommend):
  movies_ind = tf.nn.top_k(all_users_ratings_new[user_index], num_to_recommend)[1]
  return tf.gather_nd(movies, tf.expand_dims(movies_ind, axis = 1))
```

```python
user_topmovies = {}
num_to_recommend = tf.reduce_sum(tf.cast(tf.equal(users_movies, tf.zeros_like(users_movies)), dtype = tf.float32), axis = 1)
for ind in range(num_users):
  top_movies = find_user_top_movies(ind, tf.cast(num_to_recommend[ind], dtype = tf.int32))
  user_topmovies[users[ind]] = list(top_movies)
```
